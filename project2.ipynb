```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv("mission_launches.csv")
df
```

```python
df = df.replace("Arm??e de l'Air", "Armée de l'Air")
# Fixes the missing accent on the name of the organization 'Armée de l'Air.
```

```python
df['Country'] = df['Location'].apply(lambda x: x.split(',')[-1])
df
# Create a column called "Country" that extracts the country from the "Location" column in our 'df' by looking at the item after the last comma in the string. 
```

```python
df_clean = df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"])
df_clean
# Creates a new data frame that does not include columns from the original 'df' that bear no meaning to us.
```

```python
df_count = df_clean.groupby("Country").agg("count").reset_index()
df_count
# Creates a new dataframe with numerical data that has the number of launches grouped by the country the launch took place in.
```

```python
df_success = df_clean[df_clean["Mission_Status"] == "Success"].reset_index()
# Creates a new dataframe that containes only launches that had a "Mission_Status" value of "Success"

df_success.groupby("Country").agg('count').reset_index()
# Group the new data frame by country to find the amount of successful launches per country.
```

```python
df_count["Number of Successful Launches"] = [3, 2, 0, 243, 285, 2, 63, 4, 9, 113, 608, 9, 0, 11, 2, 0, 33, 1303, 1, 1, 1186, 1]
# Using the dataframe above, creates a new column that contains the amount of SUCCESSFUL launches per country in the dataframe that counts the amount of launches per country.
# Countries that did not show up in "df_success" were assigned values of 0 in this new column since they had 0 successful launches.

df_count["Number of Launches"] = df_count["Location"]
# Creates a new column that has the amount of launches per country by looking at the aggregated count of launches per location

df_count
```

```python
df_count["Success Rate"] = [(3/6)*100,(2/3)*100,(0/3)*100,(243/268)*100,(285/303)*100,(2/2)*100,(63/76)*100,(4/13)*100,(9/11)*100,(113/126)*100,(608/701)*100,(9/9)*100,(0/4)*100,(11/13)*100,(2/5)*100,(0/1)*100,(33/36)*100,(1303/1395)*100,(1/1)*100,(1/3)*100,(1186/1344)*100,(1/1)*100]
df_final = df_count.drop(columns=["Organisation", "Location","Date","Detail","Rocket_Status","Price","Mission_Status"])
df_final
```

```python
df_count.plot.bar(x = "Country", y = ["Number of Successful Launches", "Number of Launches"], log = True, width=0.75, colormap="jet")
# Creates a bar plot, the y-axis is on a log scale

plt.legend(["Number of Launches", "Number of Successful Launches"], fontsize="10", loc ="upper left")
# Adjusting the location and size of legend

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 10)
# Changes the size of the graph

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
# Changes the font size of the labels on the x and y-axes

plt.ylabel("Log(Number of Launches)", fontsize=11)
plt.xlabel("Country", fontsize=11)
plt.title("Amount of Space Mission Launches per Country", fontsize=12)
# Changes what the title, x, and y-axes labels say
```
